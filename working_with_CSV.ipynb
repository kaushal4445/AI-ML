```python
# Convertors

def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL.csv', converters={'team1' : rename})
```
