# **Making Empty Date Frame**

```python
df = pd.DataFrame()
```

```python
for i in range(1,124):
  response = requests.get('https://api.themoviedb.org/3/tv/top_rated?api_key=9b473644c9294c3a7186c778c83440c7&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(response.json()['results'])[['id', 'name','overview', 'first_air_date', 'popularity', 'vote_average', 'vote_count']]
  df = pd.concat([df, temp_df], ignore_index=True)
```
